```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("Sine Waveform")
plt.plot(x, y, color = "b")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```
